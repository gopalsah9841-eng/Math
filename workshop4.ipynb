```python
#f.A
import matplotlib.pyplot as plt
x = [123, 230, 78, 193, 12]
colors = ["Red", "Silver", "Blue", "White", "Green"]
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(x, labels=colors, explode=myexplode)
plt.title("Car Color Count")
plt.show()
```

```python
#f.B
import matplotlib.pyplot as plt
x = [123, 230, 78, 193, 12]
colors = ["Red", "Silver", "Blue", "White", "Green"]
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(x, labels=colors, explode=myexplode, shadow=True)
plt.title("Car Color Count")
plt.show()
```

```python
#f.C
import matplotlib.pyplot as plt
x = [123, 230, 78, 193, 12]
colors = ["Red", "Silver", "Blue", "White", "Green"]
plt.pie(x, labels=colors)
plt.legend(title="Car Colours:")
plt.show()
```
