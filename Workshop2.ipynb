1.Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result


print(unique_list([1,2,3,3,3,3,4,5]))
```

```python
# Exercise - 2
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
print(is_prime(7))
```

```python
# Exercise - 3
def is_prime(n): # Using the same function from exercise - 2
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
for i in range(3, 1000, 2): # Going from 3 to 1000 while jumping 2 steps
  if is_prime(i) and is_prime(i + 2): # Checking if i and i + 2 are prime
    print(i, i + 2)
```

```python
# Exercise  - 4
def sumOfDivisors(n):
    total = 0

    for i in range(1, n):
        if n % i == 0: # Checking if the i is the divisor
            total += i

    return total
print(sumOfDivisors(12))
```

```python
# Exercise - 5
def sumOfDivisors(n): # Using the same function from exercise - 4
    total = 0

    for i in range(1, n):
        if n % i == 0:
            total += i

    return total
n = 7 # Going from 7 because we need to find the perfect number bigger than 6
while True:
  if sumOfDivisors(n) == n: # Perfect number condition
    print("The next perfect number is: ", n)
    break
  n += 1
```

```python
# Exercise - 6
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for m in marks:
    total += m
mean = total / len(marks)

# Median
n = len(marks)
if n % 2 == 1:
    median = marks[n // 2] # For even number of data
else:
    median = (marks[n // 2 - 1] + marks[n // 2]) / 2 # For odd number of data

print("Mean --> ", mean)
print("Median --> ", median)
```

```python
def convert_to_binaryNum(n):
    binary = ""  # Initialising an empty string for binary

    while n > 0:
        binary += str(n % 2)  # Store remainder
        n = n // 2            # Update n

    return binary[::-1]  # Reverse string

print(convert_to_binaryNum(11))
```

```python
# Exercise - 8
def prodDigits(n):
  prod = 1 # Initialising to 1 because we have to multiply
  while n > 0:
    prod *= n % 10 # Multiplying with the remainder
    n = n // 10
  return prod
print(prodDigits(27))
```

```python
# Exercise - 9
def multiply(*args): # *args gets multiple arguments
  prod = 1 # Initialising to 1 because we have to multiply
  for i in args:
    prod *= i # Multiplying every argument
  return prod
print(multiply(1,2,3,4,5))
print(multiply(12, 13, 14))
```
